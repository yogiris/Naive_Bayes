```python
trainfordum = trainset.iloc[:,0:13]
trainfordum = pd.get_dummies(trainfordum)
trainfordum
```

```python
trainwithdum = pd.concat([trainfordum, trainset['Salary']],axis=1)
trainwithdum
```

```python
testfordum = testset.iloc[:,0:13]
testfordum = pd.get_dummies(testfordum)
testfordum
```

```python
testwithdum = pd.concat([testfordum, testset['Salary']],axis=1)
testwithdum
```

```python
X = trainwithdum.values[:,0:102]
Y = trainwithdum.values[:,102]
x = testwithdum.values[:,0:102]
y = testwithdum.values[:,102]
```

```python
# Multinomial Naive Bayes
clf_mb = MB()
clf_mb.fit(X,Y)
train_pred_multi = clf_mb.predict(X)
accuracy_train_multi = np.mean(train_pred_multi==Y)
test_pred_multi = clf_mb.predict(x)
accuracy_test_multi = np.mean(test_pred_multi==y)

print('Training accuracy of Multinomial Naive Bayes is:',accuracy_train_multi,'\n',
         'Testing accuracy of Multinomial Naive Bayes is:',accuracy_test_multi)
```

```python
# Gaussian Naive Bayes 
clf_gb = GB()
clf_gb.fit(X,Y) 
train_pred_gaus = clf_gb.predict(X)
accuracy_train_gaus = np.mean(train_pred_gaus==Y)

test_pred_gaus = clf_gb.predict(x)
accuracy_test_gaus = np.mean(test_pred_gaus==y)

print('Training accuracy of Gaussian Naive Bayes is:',accuracy_train_gaus,'\n',
        'Testing accuracy of Gaussian Naive Bayes is:',accuracy_test_gaus)
```

```python
#From the above models, it is clear that the Gaussian naive bayes model gives a higher accuracy in classifying the salary data
```
